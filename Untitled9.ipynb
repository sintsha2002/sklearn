```python
# Install
!pip install xgboost -q

# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Mount Google Drive
drive.mount('/content/drive')

# Read CSV file
file_path = "/content/drive/MyDrive/Woreta daily rainfall & temperature data.csv"

df = pd.read_csv(file_path, header=None)

print("Original shape:", df.shape)
print(df.head())

# Correct column names
columns = (
    ["station_name", "station_code", "latitude", "longitude", "elevation",
     "element", "year", "month", "missing_flag"]
    + [f"day_{i:02d}" for i in range(1, df.shape[1] - 9 + 1)]
)

df.columns = columns

print("Columns:")
print(df.columns.tolist())

# Keep rainfall records only
rainfall_elements = ["PRECIP", "RAIN", "RAINFALL"]

df = df[df["element"].astype(str).str.upper().isin(rainfall_elements)]

# Convert monthly wide data to daily time-series data
records = []

day_columns = [col for col in df.columns if col.startswith("day_")]

for _, row in df.iterrows():
    year = int(row["year"])
    month = int(row["month"])
    station = row["station_name"]

    for col in day_columns:
        day = int(col.split("_")[1])
        rainfall = row[col]

        try:
            date = pd.to_datetime(f"{year}-{month}-{day}")
            records.append([date, station, rainfall])
        except:
            pass

rain_df = pd.DataFrame(records, columns=["date", "station_id", "rainfall"])

rain_df["rainfall"] = pd.to_numeric(rain_df["rainfall"], errors="coerce")
rain_df.dropna(inplace=True)
rain_df.sort_values("date", inplace=True)
rain_df.reset_index(drop=True, inplace=True)

print("Daily rainfall records:", len(rain_df))
print(rain_df.head())

# Feature engineering
rain_df["year"] = rain_df["date"].dt.year
rain_df["month"] = rain_df["date"].dt.month
rain_df["day"] = rain_df["date"].dt.day
rain_df["day_of_year"] = rain_df["date"].dt.dayofyear

rain_df["lag1"] = rain_df["rainfall"].shift(1)
rain_df["lag2"] = rain_df["rainfall"].shift(2)
rain_df["lag3"] = rain_df["rainfall"].shift(3)
rain_df["lag7"] = rain_df["rainfall"].shift(7)

rain_df["roll3"] = rain_df["rainfall"].rolling(3).mean()
rain_df["roll7"] = rain_df["rainfall"].rolling(7).mean()
rain_df["roll30"] = rain_df["rainfall"].rolling(30).mean()

rain_df.dropna(inplace=True)

# =========================
# XGBOOST MODEL
# =========================

features = [
    "year", "month", "day", "day_of_year",
    "lag1", "lag2", "lag3", "lag7",
    "roll3", "roll7", "roll30"
]

X = rain_df[features]
y = rain_df["rainfall"]

split = int(len(X) * 0.8)

X_train = X.iloc[:split]
X_test = X.iloc[split:]

y_train = y.iloc[:split]
y_test = y.iloc[split:]

xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    objective="reg:squarederror",
    random_state=42
)

xgb.fit(X_train, y_train)

xgb_pred = xgb.predict(X_test)

xgb_mae = mean_absolute_error(y_test, xgb_pred)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
xgb_r2 = r2_score(y_test, xgb_pred)

print("\nXGBoost Results")
print("MAE :", xgb_mae)
print("RMSE:", xgb_rmse)
print("R2  :", xgb_r2)

# =========================
# LSTM MODEL
# =========================

rain_series = rain_df[["rainfall"]].values

scaler = MinMaxScaler()
rain_scaled = scaler.fit_transform(rain_series)

TIME_STEPS = 30

def create_dataset(data, time_steps):
    X, y = [], []

    for i in range(len(data) - time_steps):
        X.append(data[i:i + time_steps])
        y.append(data[i + time_steps])

    return np.array(X), np.array(y)

X_lstm, y_lstm = create_dataset(rain_scaled, TIME_STEPS)

split = int(len(X_lstm) * 0.8)

X_train_lstm = X_lstm[:split]
X_test_lstm = X_lstm[split:]

y_train_lstm = y_lstm[:split]
y_test_lstm = y_lstm[split:]

model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(TIME_STEPS, 1)))
model.add(Dropout(0.2))
model.add(LSTM(32))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")

history = model.fit(
    X_train_lstm,
    y_train_lstm,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

lstm_pred_scaled = model.predict(X_test_lstm)

lstm_pred = scaler.inverse_transform(lstm_pred_scaled)
y_actual = scaler.inverse_transform(y_test_lstm)

lstm_mae = mean_absolute_error(y_actual, lstm_pred)
lstm_rmse = np.sqrt(mean_squared_error(y_actual, lstm_pred))
lstm_r2 = r2_score(y_actual, lstm_pred)

print("\nLSTM Results")
print("MAE :", lstm_mae)
print("RMSE:", lstm_rmse)
print("R2  :", lstm_r2)

# =========================
# MODEL COMPARISON
# =========================

results = pd.DataFrame({
    "Model": ["XGBoost", "LSTM"],
    "MAE": [xgb_mae, lstm_mae],
    "RMSE": [xgb_rmse, lstm_rmse],
    "R2": [xgb_r2, lstm_r2]
})

print("\nModel Comparison")
print(results)

# =========================
# PLOTS
# =========================

plt.figure(figsize=(14,5))
plt.plot(y_test.values, label="Actual Rainfall")
plt.plot(xgb_pred, label="XGBoost Prediction")
plt.title("XGBoost Rainfall Forecasting")
plt.xlabel("Time")
plt.ylabel("Rainfall")
plt.legend()
plt.show()

plt.figure(figsize=(14,5))
plt.plot(y_actual, label="Actual Rainfall")
plt.plot(lstm_pred, label="LSTM Prediction")
plt.title("LSTM Rainfall Forecasting")
plt.xlabel("Time")
plt.ylabel("Rainfall")
plt.legend()
plt.show()
```
